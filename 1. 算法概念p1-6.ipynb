## 时间复杂度

比较左边四组代码哪组运行时间最短？

用什么方式来比较算法的快慢？<br>
1. 若用时间比较，不同机器运行的速度不同，无法比较。
2. n为问题的规模，for循环执行n次，表示问题的复杂度、问题的规模为n。

### 例子 

下面例子设n=100

O(1)解释: O表示估计，1表示单位为1

1. 时间复杂度定义为 O(1)

```python
import time
start = time.time()
a=0
end = time.time()
print('Running time: %s ms'%((end-start)/1000))
```

2. 时间复杂度定义为 O(n)

```python
start = time.time()
a=0
for i in range(100):
    a+=1
end = time.time()
print('Running time: %s ms'%((end-start)/1000),a)
```

3. 时间复杂度为O(n^2)

```python
start = time.time()
a=0
for i in range(100):
    for j in range(100):
        a+=1
end = time.time()
print('Running time: %s Seconds'%((end-start)/1000),a)
```

4. 时间复杂度为O(n^3)

```python
start = time.time()
a=0
for i in range(100):
    for j in range(100):
        for k in range(100):
            a+=1
end = time.time()
print('Running time: %s Seconds'%((end-start)/1000),a)
```

5. 时间复杂度为 O(3)×，O(n)中n为单位，所以此处应该是1个单位-->O(1)<br>
此处表示近似，非精确，即只要问题规模不到n，就算作O(1)

```python
print('Hello world')
print('hello python')
print('hello algorithm')
```

6. 时间复杂度为 O(n^2+n)×,同理-->O(n^2)

```python
n = 100
for i in range(n):
    print('hello world')
    for j in range(n):
        print('hello world')
```

7. 时间复杂度计算:<br>
$2^6 = 64$<br>
$log_2^{64}=6$<br>
因此，定义时间复杂度为 $O(log_2^n)$ 或 $O(logn)$ <br>
<b>当程序每次循环，规模减半时，则复杂度同上</b>

```python
n = 64
while n > 1:
    print(n)
    n = n//2
```

### 时间复杂度小结 

1. 时间复杂度是用来估计算法运行时间的一个式子（单位）。
2. 一般来说，时间复杂度高的算法比复杂度低的算法慢。
3. 常见的时间复杂度（按照效率排序）: <br> O(1) < O(logn) < O(n) < $O(nlogn)$ < $O(n^2)$ < $O(n^2logn)$ < $O(n^3)$ <br>
4. 复杂问题的时间复杂度：<br> $O(n!)$, $O(n^2)$, $O(n^n)$

### 如何简单快速地判断算法复杂度

大多数情况下：<br>
1. 确定问题规模n<br>
2. 循环减半过程-->logn<br>
3. k层关于n的循环-->$n^k$<br>

复杂情况：根据算法执行过程判断

## 空间复杂度 

空间复杂度：用来评估算法内存占用大小的式子<br>
空间复杂度的表达方式与时间复杂度完全一样<br>
1. 算法使用了几个变量：O(1)
2. 算法使用了长度为n的一维列表：O(n)
3. 算法使用了m行n列的二维列表：O(mn)<br>

<b>“空间换时间”</b>，时间比空间重要。such as分布式运算。

## 递归

递归的两个特点：
1. 调用自身
2. 结束条件

```python
# 不是合法的递归，因为没有结束条件
def func1(x):
    print(x)
    func1(x-1)
```

```python
# 不是合法的递归，若x起始大于0，则没有结束条件
def func2(x):
    if x>0:
        print(x)
        func2(x+1)
```

```python
# 是合法的递归
def func3(x):
    if x>0:
        print(x)
        func3(x-1)
```

```python
# 是合法的递归
def func4(x):
    if x>0:
        func4(x-1)
        print(x)
```

```python
func3(3)
```

```python
func4(3)
```

### 例：汉诺塔问题 

1. 在一块铜板装置上，有三根杆(编号A、B、C)，在A杆自下而上、由大到小按顺序放置64个金盘。
2. 把A杆上的金盘全部移到C杆上，并仍保持原有顺序叠好。
3. 每次只能移动一个盘子，并且在移动过程中三根杆上都始终保持大盘在下，小盘在上，操作过程中盘子可以置于A、B、C任一杆上。

n=2时，
1. 把小圆盘从A移动到B
2. 把大圆盘从A移动到C
3. 把小圆盘从B移动到C

n>2时，<br>
把上面n-1个圆盘看作一个整体，把底下最大的一个圆盘看作一个
1. 把n-1个圆盘从A经过C移动到B
2. 把第n个圆盘（最底下的）从A移动到C
3. 把n-1个小圆盘从B经过A移动到C

```python
# n个盘子,a表示初始存圆盘的位置，b表示圆盘经过的位置，c表示圆盘最终的位置
def hanoi(n, a, b, c):
    if n > 0:
        hanoi(n-1, a, c, b) #step1
        print('moving from %s to %s'%(a,c)) #step2
        hanoi(n-1, b, a, c) #step3
        
hanoi(3, "A",'B','C')
```

假设n个圆盘移动需要h(n)次：<br>
n个盘子，需要 $h(n-1)+1+h(n-1)$ 步，则 $h(n) = 2*h(n-1)+1$<br>
可算出 $h(n) ~= 2^n$
