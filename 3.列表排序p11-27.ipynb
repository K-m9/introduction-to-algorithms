### 冒泡排序(Bubble Sort)

1. 列表每两个相邻的数，如果前面比后面大，则交换这两个数。
2. 一趟排序完成后，则无序区减少一个数，有序区增加一个数(最大或最小的一个数)。
3. 一共需要循环n-1次，n=len(list)。
4. 代码关键点：趟（从第0趟开始）、无序区范围(0,n-i)。

<b>时间复杂度: $O(n^2)$ ，算法中没有折半，有两层</b>

```python
import random
```

```python
# 升序
def bubble_sort(li):
    for i in range(len(li)-1): #第i趟
        for j in range(len(li)-i-1):
            if li[j] > li[j+1]:
                li[j], li[j+1] = li[j+1], li[j]
        print(li)
```

```python
li = [random.randint(0,1000) for i in range(1000)]
print(li)
```

```python
bubble_sort(li)
print(li)
```

#### 改进的冒泡排序 

若第i趟排序之后，已完成按顺序排序，则不用继续进行排序。

```python
# 升序
@cal_time
def bubble_sort(li):
    for i in range(len(li)-1): #第i趟
        exchange = False
        for j in range(len(li)-i-1):
            if li[j] > li[j+1]:
                li[j], li[j+1] = li[j+1], li[j] #每趟把最大的放到最后
                exchange = True #第i趟遍历无需要更换顺序的数字
        #print(li)
        if not exchange:
            return
```

### 选择排序(Select Sort)

简单版：<br>
1. 先遍历一遍列表，找到最小的值，取出
2. 剩下的再遍历一遍，找到第二小的值，取出<br>
...

```python
def select_sort_simple(li):
    li_new = []
    for i in range(len(li)):
        min_val = min(li) #需要遍历一遍，时间复杂度O(n)
        li_new.append(min_val)
        li.remove(min_val) #删除后每个数字需往前挪一位，时间复杂度O(n)
    return li_new            
```

```python
li = [3,2,5,4,6,2,1,2]
select_sort_simple(li)
```

select_sort_simple缺点：<br>
1. 生成了两个列表，占用了两份内存；
2. 时间复杂度为 $O(n^2)$

改进：<br>
将列表分为有序区（假定为最左端）和无序区，<br>
1. 一趟排序记录最小的数，放到第一个位置，有序区空间+1;
2. 再一趟排序记录列表无序区最小的数，放到第二个位置（将无序区最小值与无序区第一个位置的数交换），有序区空间+1;<br>
...
3. 算法关键点：有序区和无序区、无序区最小数的位置

```python
def select_sort(li):
    for i in range(len(li)-1): # i是第几趟
        min_loc = i #记为无序区的第一个位置
        for j in range(i+1, len(li)): #不需要自己和自己比
            if li[j] < li[min_loc]:
                #li[j], li[min_loc] = li[min_loc], li[j]
                min_loc = j #更新最小的下标
        li[i], li[min_loc] = li[min_loc], li[i]
    return li
```

```python
li = [3,2,5,4,6,2,1,2]
select_sort(li)
```

```python
def select_sort(li):
    for i in range(len(li)-1):
        min_loc = i
        for j in range(i+1, len(li)):
            if li[j] < li[min_loc]:
                min_loc = j
        if min_loc != i:
            li[i], li[min_loc] = li[min_loc], li[i]
    return li
```

时间复杂度：$O(n^2)$

### 插入排序(Insert Sort)

1. 初始时手里（有序区）只有一张牌
2. 每次（从无序区）摸一张牌，插入到手里已有排的正确位置

需要摸n-1张牌，对有序区进行排序插入（从右往左排序，若取得的牌小于有序区最右的数，最右的数往后移一位，以此类推）

```python
def insert_sort(li):
    for i in range(1, len(li)): # i 表示摸到的牌的下标
        tmp = li[i]
        j = i - 1 # j 指手里的牌的下标（有序区）
        while li[j] > tmp and j >= 0:
            li[j+1] = li[j]
            j -= 1
        li[j+1] = tmp
        #print(li)
    return li
```

```python
li = [3,2,6,5,4,2,1,6]
```

```python
insert_sort(li)
```

时间复杂度：<br>
摸第1趟1次排序，第2趟2次排序，第n趟n次排序，加总 $n^2/2$<br>
即 $O(n^2)$

### 快速排序 

1. 快速排序：快
2. 快速排序思路：
    - 取一个元素p（第一个元素），使元素p归位；
    - 列表被p分成两部分，左边都p小，右边都比p大；
    - 递归完成排序。(左右两部分分别排序)

```python
def partition(li, left, right):
    tmp = li[left]
    while left < right:
        while left < right and li[right] >= tmp: #从右边找比tmp小的数
            right -= 1   #往左走
        li[left] = li[right] #把右边的值写到左边的空位
        while left < right and li[left] <= tmp:
            left += 1
        li[right] = li[left] #把左边的值写到右边的空位
    li[left] = tmp #把tmp值归位
    return left #返回mid值
```

```python
@cal_time
def quick_sort(data, left, right):
    if left < right:
        mid = partition(data, left, right) #中间归位
        quick_sort(data, left, mid - 1) #左边排序
        quick_sort(data, mid + 1, right) #右边排序
    return data
```

```python
li = [5,7,6,8,1,2,3]
partition(li, 0 ,len(li)-1)
```

```python
li = [5,7,6,8,1,2,3]
quick_sort(li, 0, len(li)-1)
```

```python
def _quick_sort(li, left, right):
    if left < right:
        mid = partition(li, left, right)
        _quick_sort(li, left, mid-1)
        _quick_sort(li, mid+1, right)
```

```python
@cal_time
def quick_sort(li):
    _quick_sort(li, 0, len(li)-1)
```

#### 快速排序存在的问题 

1. 快速排序的效率：
    - 快速排序的时间复杂度

每次运行都为O(n)，但每次运行都分两半，一共分$O(logn)$ 次，
则时间复杂度为 $O(nlogn)$

2. 快速排序的问题：
    - 最坏情况
    - 递归

递归：存在最大深度问题，消耗系统内存。<br>

最坏情况：每次递归只移动一个数，时间复杂度相当于$O(n^2)$

```python
def partition(li, left, right):
    tmp = li[left]
    while left < right:
        while left < right and li[right] >= tmp: #从右边找比tmp小的数
            right -= 1   #往左走
        li[left] = li[right] #把右边的值写到左边的空位
        #print(li)
        while left < right and li[left] <= tmp:
            left += 1
        li[right] = li[left] #把左边的值写到右边的空位
        #print(li)
    li[left] = tmp #把tmp值归位
    return left #返回mid值
```

### 堆排序 

1. 树是一种数据结构，如：目录结构
2. 树是一种可以递归定义的数据结构
3. 数是由n个节点组成的集合：
    - 如果n=0，那这是一棵空树；
    - 如果n>0，那存在1个节点作为树的根节点，其他节点可以分为m个集合，每个集合本身又是一棵树。

一些概念：
- 根节点：最顶端的节点（一个）；叶子节点：最后一层的节点（不能往下分）
- 树的深度（高度）：一共多少层
- 树的度：节点有多少个子节点即为多少度
- 孩子节点；父节点
- 子树：分叉下面的子树

#### 二叉树概念 

- 二叉树：度不超过2的树
- 每个节点最多有两个孩子节点
- 两个孩子节点被区分为左孩子节点和右孩子节点

- 满二叉树：一个二叉树，如果每一层的结点数都达到最大值，则这个二叉树就是满二叉树。
- 完全二叉树：叶节点只能出现在最下层和次下层，并且最下面一层的结点都集中在该层最左边的若干位置的二叉树。（除了最下面一层，其他层满，最下面一层集中在左边）

#### 堆排序前传--二叉树的存储方式 

二叉树的存储方式（表示方式）<br>
- 链式存储方式
- 顺序存储方式
    - 父节点和左孩子结点的编号下标有什么关系？
        - 0-1 1-3 2-5 3-7 4-9
        - i --> 2i+1
    - 父节点和右孩子结点的编号下标有什么关系？
        - 0-2 1-4 2-6 3-8 4-10
        - i --> 2i+2

#### 堆排序概念 

堆：一种特殊的完全二叉树结构
- 大根堆：一棵完全二叉树，满足任一节点都比其孩子节点大
- 小根堆：一棵完全二叉树，满足任一节点都比其孩子节点小

堆的向下调整性质：
- 假设根节点的左右子树都是堆，但<b>根节点</b>自身不满足堆的性质
- 可以通过一次向下的调整来将其变成一个堆。(把根节点往下放)

堆排序过程：
1. 建立堆：先建立小堆，再扩展堆，获得最终堆
2. 得到堆顶元素，为最大元素
3. 去掉堆顶，将堆最后一个元素放到堆顶，此时可通过一次调整重新使堆有序
4. 堆顶元素为第二大元素
5. 重复步骤三，直到堆变空

```python
def sift(li, low, high):
    '''
    li: 列表
    low: 堆的根节点
    high: 堆的最后一个元素的位置
    '''
    i = low
    j = 2 * i + 1 # j开始是左孩子节点
    tmp = li[low] # 把堆顶存起来
    while j <= high: # 只要j位置有数
        if j+1 <= high and li[j+1] > li[j]: # 如果右孩子有并且比较大
            j = j + 1 # j指向右孩子
        if li[j] > tmp:
            li[i] = li[j]
            i = j # 往下看一层
            j = 2 * i + 1
        else:     # tmp更大，把tmp放到i的位置上
            li[i] = tmp # 把tmp放到某一级领导位置上
            break
    else:
        li[i] = tmp #把tmp放到叶子节点上
```

```python
def heap_sort(li):
    n = len(li)
    for i in range((n-2)//2 - 1, -1, -1):
        # i表示建堆的时候调整的部分的根的下标
        sift(li, i, n-1)
    # 建堆完成了
    for i in range(n-1, -1, -1):
        # i 指向当前堆的最后一个元素
        li[0], li[i] = li[i], li[0]
        sift(li, 0, i - 1) # i-1是新的high 
```

```python
li = [i for i in range(100)]
import random
random.shuffle(li)
```

```python
heap_sort(li)
```
